```python
from datetime import datetime
import instaloader
# import json
import pandas as pd 
import emoji

account_my = 'myaccount';
account_target = 'jokowi';
# account_target = 'kemenhub151';
L = instaloader.Instaloader()
L.interactive_login(account_my)      # (ask password on terminal)


posts = instaloader.Profile.from_username(L.context, account_target).get_posts()
SINCE = datetime(2021, 4, 1)  # further from today, inclusive
UNTIL = datetime(2021, 5, 12)  # closer to today, not inclusive

i_all = 134
i = 5
k = 0  # initiate k
#k_list = []  # uncomment this to tune k

print('start!')
for post in posts:
    postdate = post.date

    if postdate > UNTIL:
        continue
    elif postdate <= SINCE:
        k += 1
        if k == 50:
            break
        else:
            continue
    else:
        i_all = i_all + 1
        print('\n\npost-'+str(i_all)+' -- '+str(postdate))
#         print(post.caption)
        
        if post.caption.find("mudik") != -1:
            print('_____________________________________________________"MUDIK" included')
            i = i+1
            filename = account_target+"_post_"+str(i)+"_"+str(post.mediaid)
            
            textfile = open("result/"+filename+"_post.txt", 'w', encoding='utf-8')
            textfile.write(post.caption)
            textfile.close()
            L.download_post(post, "postattr__"+filename)

            comments_from_loop_including_answers = []
            for comment in post.get_comments():
                comments_from_loop_including_answers.append(comment)
                for answer in comment.answers:
                    comments_from_loop_including_answers.append(answer)
            pd.DataFrame(comments_from_loop_including_answers).to_csv("bisa/"+filename+"_comment.csv")
#             a = numpy.asarray(comments_from_loop_including_answers)            
#             numpy.savetxt(filename, a, delimiter=",")
#             print(comments_from_loop_including_answers)
        
        # if you want to tune k, uncomment below to get your k max
        #k_list.append(k)
        k = 0  # set k to 0

#max(k_list)

print('\n\n....done')
```
